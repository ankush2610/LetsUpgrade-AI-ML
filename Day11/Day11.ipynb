```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
data=df.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python

datas=data.replace(['Yes','No'],[1,0])
datas.head()
```

```python
datas.columns
```

# Hypothesis-1 Attrition vs Age
Ho = No correlation between Attrition and age

Ha = Significant correlation between Attrition and age


```python
stats,p=pearsonr(datas.Attrition,datas.Age)
print(stats,p)
```

From the above code, 

p < 0.05. Therefore, we reject the Null Hypothesis.

So, there is Significant correlation between Attrition and age
