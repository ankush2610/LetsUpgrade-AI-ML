```python
df3.columns
```

# Hypothesis-1 Wilcoxon Sign test

Ho= There is no significant difference between JobLevel and Number of Companies Worked.

Ha= There is significant difference between JobLevel and Number of Companies Worked.

```python
stats,p=wilcoxon(df3.JobLevel,df3.NumCompaniesWorked)
print(stats,p)
```

From the above p value, since p<0.05, we can conclude that the __Alternate hypothesis is accepted__.

# Hypothesis-2 Friedman Test

Ho= There is no significant difference between JobLevel, Number of Companies Worked and Education.

Ha= There is significant difference between JobLevel, Number of Companies Worked and Education.

```python
stats,p=friedmanchisquare(df3.JobLevel,df3.NumCompaniesWorked,df3.Education)
print(stats,p)
```

From the above p value, since p<0.05, we can conclude that the __Alternate hypothesis is accepted__.

# Hypothesis-3 Mannwhitney Test

Ho= There is no significant difference between JobLevel and Number of Companies Worked.

Ha= There is significant difference between JobLevel, Number of Companies Worked.

```python
stats,p=mannwhitneyu(df3.JobLevel,df3.NumCompaniesWorked)
print(stats,p)
```

From the above p value, since p>0.05, we can conclude that the __Null hypothesis is accepted__.

# Hypothesis-4 Kruskal Wallis Test

Ho= There is no significant difference between JobLevel, Number of Companies Worked and Education.

Ha= There is significant difference between JobLevel, Number of Companies Worked and Education.

```python
stats,p=kruskal(df3.TotalWorkingYears,df3.NumCompaniesWorked,df3.TrainingTimesLastYear)
print(stats,p)
```

From the above p value, since p<0.05, we can conclude that the __Alternate hypothesis is accepted__.

# Hypothesis-5 Chi-Square Test

Using Chi-Square Test, lets check the dependency between Gender and Department.

```python
chisquare=pd.crosstab(df1.Attrition,df1.Gender)
chisquare
```

```python
stats,p,dof,expected=chi2_contingency(chisquare)
print(stats,p)
```

From the above p value, since p>0.05, we can conclude that the __Null hypothesis is accepted__.

__There is no dependency between Gender and Attrition__.
