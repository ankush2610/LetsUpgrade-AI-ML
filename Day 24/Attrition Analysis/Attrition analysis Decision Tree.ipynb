# Filling Null Values 

```python
data['TotalWorkingYears'] = data['TotalWorkingYears'].fillna(11.28) 
# 11.28 is the mean of TotalWorkingYears column 

data['NumCompaniesWorked'] = data['NumCompaniesWorked'].fillna(2)
# 2 is the median of NumCompaniesWorked
```

```python
data['Attrition']=preprocessing.LabelEncoder().fit_transform(data['Attrition'])
data['BusinessTravel']=preprocessing.LabelEncoder().fit_transform(data['BusinessTravel'])
data['Department']=preprocessing.LabelEncoder().fit_transform(data['Department'])
data['EducationField']=preprocessing.LabelEncoder().fit_transform(data['EducationField'])
data['MaritalStatus']=preprocessing.LabelEncoder().fit_transform(data['MaritalStatus'])
data['Gender']=preprocessing.LabelEncoder().fit_transform(data['Gender'])
data['JobRole']=preprocessing.LabelEncoder().fit_transform(data['JobRole'])
data.head()
```
