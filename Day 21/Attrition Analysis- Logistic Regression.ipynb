Let us perform Logistic Regression on Attrition Analysis Dataset. 

Here, the traget variable is __Attrition__ which is binary categorical.

```python
data['TotalWorkingYears']=data['TotalWorkingYears'].fillna(data['TotalWorkingYears'].mean())
data['NumCompaniesWorked']=data['NumCompaniesWorked'].fillna(data['NumCompaniesWorked'].median())
```

```python
data.isnull().sum()
```

```python
data = data.drop(['ID','EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()

data['Attrition'] = lab.fit_transform(data['Attrition'])
data['BusinessTravel'] = lab.fit_transform(data['BusinessTravel'])
data['Department'] = lab.fit_transform(data['Department'])
data['MaritalStatus'] = lab.fit_transform(data['MaritalStatus'])
data['JobRole'] = lab.fit_transform(data['JobRole'])
data['Gender'] = lab.fit_transform(data['Gender'])
data['EducationField'] = lab.fit_transform(data['EducationField'])
```

```python
data.head()
```

```python
data.shape
```

```python
y=data[['Attrition']]
x=data.drop(['Attrition'],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_test)
ypred[0:5]
```

```python
y_test.head()
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypred)
```
