# Import useful libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix, accuracy_score
```

# Reading the Titanic Training Dataset

```python
df = pd.read_csv('C:/Users/Vishal/Desktop/LU aiml/Data/Titanic training dataset.csv')
df.head()
```

```python
lab = LabelEncoder()

df["Sex"] = lab.fit_transform(df["Sex"])
df["Embarked"] = lab.fit_transform(df["Embarked"])
```

```python
df.head()
```

# Checking Null Values

```python
df.isnull().sum()
```

# dropping irrelevent columns

```python
df1=df.drop(['Cabin','PassengerId','Name','Ticket'],axis=1)
df1.head()
```

```python
def prediction(col):
    
    clf = svm.SVC(gamma = 0.01, C = 100)
    acc=[]
    for i in col:
        x = df1.drop([i], axis=1)
        print("====================================================================")
        print(f"The Dependent Variable is {i} & remaining are Independent Variables")
        y = df1[i]
    
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state=0)
        support = clf.fit(x_train,y_train)
        ypred = support.predict(x_test)
        scored=accuracy_score(y_test, ypred, normalize=True)
        print(f"\n Acuuracy Score for {i} is: ", accuracy_score(y_test, ypred, normalize=True))
        acc.append(scored)
        print(f"\n Confusion matrix for {i} is \n", confusion_matrix(y_test, ypred))
    return acc
```

```python
c = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch','Embarked']        
prediction(c)
```

# Inference:

The Pclass as a dependent variable and remaining as Independent variable has the highest accuracy score.
